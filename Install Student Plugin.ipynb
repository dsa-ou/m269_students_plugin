This will get the plugin and put it in the right place.

```python
import subprocess
import requests
import os

command = "pip show jupyter_contrib_nbextensions"
process = subprocess.Popen(command.split(), stdout=subprocess.PIPE)
output, error = process.communicate()
results = output.decode("utf-8").split()
getloc = False
for line in results:
    if getloc == True:
        location = line
        getloc = False
    if line == 'Location:':
        getloc = True
location += '/jupyter_contrib_nbextensions/nbextensions'
print('Nbextension location identified:',location)
print('Making directory')
path = os.path.join(location, 'm269_student')
try:
    os.mkdir(path)
except:
    print('Path already exists. This will upgrade.')
urls = {'main.js':'https://raw.githubusercontent.com/dsa-ou/m269_students_plugin/main/m269_student/main.js',
        'description.yaml':'https://raw.githubusercontent.com/dsa-ou/m269_students_plugin/main/m269_student/description.yaml',
        'readme.md':'https://raw.githubusercontent.com/dsa-ou/m269_students_plugin/main/m269_student/readme.md'}
for filename in urls:
    print('Getting',filename,'from',urls[filename])
    open(os.path.join(path, filename), 'w').write(requests.get(urls[filename], allow_redirects=True).content.decode("utf-8"))

```

This will install the plugin into nbextension.

Try this first:

```python
!jupyter contrib nbextensions install
```

If it gives you a permission error, try this instead:

```python
!jupyter contrib nbextensions install --user
```
